```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('../datasets/Mall_Customers.csv')
X = dataset.iloc[: , [3,4]].values
```

```python
from sklearn.cluster import KMeans
# elbow method
wcss=[]
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42, n_init='auto')
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss, color='purple')
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kmeans=KMeans(n_clusters=5, init='k-means++', random_state=42, n_init='auto')
# returns the dependent variable or clusters named 0-4
y_means=kmeans.fit_predict(X)
```

```python
print(y_means)
```

```python
plt.scatter(X[y_means==0,0], X[y_means==0,1], s=100, c='red', label='cluster1')
plt.scatter(X[y_means==1,0], X[y_means==1,1], s=100, c='blue', label='cluster2')
plt.scatter(X[y_means==2,0], X[y_means==2,1], s=100, c='green', label='cluster3')
plt.scatter(X[y_means==3,0], X[y_means==3,1], s=100, c='cyan', label='cluster4')
plt.scatter(X[y_means==4,0], X[y_means==4,1], s=100, c='magenta', label='cluster5')

plt.scatter(kmeans.cluster_centers_[: ,0], kmeans.cluster_centers_[:,1], s=300,  c='yellow', label='centriods')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (K$)')
plt.ylabel('Spendings (1-100)')
plt.legend()
plt.show()
```
